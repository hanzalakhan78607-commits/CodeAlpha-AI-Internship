```python
from googletrans import Translator

translator = Translator()

print("🌍 AI Language Translator (CodeAlpha Project)")
print("Type 'exit' to stop")

languages = {
    "ur": "Urdu",
    "fr": "French",
    "es": "Spanish",
    "ar": "Arabic",
    "hi": "Hindi"
}

while True:
    text = input("\nEnter text: ")

    if text.lower() == "exit":
        print("Translator stopped 👋")
        break

    print("\nSelect Language Code:")
    for code, lang in languages.items():
        print(f"{code} -> {lang}")

    lang_code = input("\nEnter language code: ")

    try:
        result = translator.translate(text, dest=lang_code)

        print("\n========================")
        print("Original:", text)
        print("Translated:", result.text)
        print("========================")

    except:
        print("⚠ Invalid language code or error occurred")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

print("🤖 AI FAQ Chatbot (CodeAlpha Project)")
print("Type 'bye' to exit")

questions = [
    "What is AI?",
    "What is Python?",
    "What is Machine Learning?",
    "What is Data Science?",
    "What is CodeAlpha?"
]

answers = [
    "AI is Artificial Intelligence where machines simulate human intelligence.",
    "Python is a high-level programming language used in AI, web and automation.",
    "Machine Learning is a branch of AI where systems learn from data.",
    "Data Science is the field of extracting insights from data.",
    "CodeAlpha is an internship platform to learn real-world tech skills."
]

while True:
    user = input("\nYou: ")

    if user.lower() == "bye":
        print("Bot: Goodbye 👋")
        break

    data = questions + [user]

    vectorizer = CountVectorizer()
    vectors = vectorizer.fit_transform(data).toarray()

    similarity = cosine_similarity([vectors[-1]], vectors[:-1])

    index = similarity.argmax()

    if similarity[0][index] == 0:
        print("Bot: Sorry, I don't understand that question.")
    else:
        print("Bot:", answers[index])
```
